```python
# Leggi il file Parquet
df_parquet = pd.read_parquet("influxdb_query_20251020_074134.parquet")
print(f"File Parquet caricato: {df_parquet.shape[0]} righe, {df_parquet.shape[1]} colonne")
df_parquet.head(5)
```

```python
# Dividi il dataset per ogni msr_subject e salva file Parquet separati
import os

# Crea una cartella per i dataset suddivisi
output_dir = "datasets_by_subject"
os.makedirs(output_dir, exist_ok=True)

# Estrai i valori unici di msr_subject
subjects = df_parquet['msr_subject'].unique()
print(f"Trovati {len(subjects)} msr_subject:")
print(subjects)

# Per ogni subject, crea un file Parquet separato
for subject in subjects:
    df_subject = df_parquet[df_parquet['msr_subject'] == subject]
    filename = f"{output_dir}/{subject}.parquet"
    df_subject.to_parquet(filename, compression='snappy', engine='fastparquet')
    print(f"✓ Salvato: {filename} ({df_subject.shape[0]} righe)")
```
